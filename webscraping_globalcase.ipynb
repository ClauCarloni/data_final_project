- Going to the specific site

```python
# function to check number of months between first and last date
from datetime import datetime

def days_between(d1, d2):
    d1 = datetime.strptime(d1, '%Y-%m')
    d2 = datetime.strptime(d2, '%Y-%m')
    return round(abs((d2 - d1).days/30))
```

```python
# months between 2014-02 and 2021-10
days_between('2004-02', '2021-10')
```

```python
# creating the dates list to run across
dates = []
start = '2004-02'
next = ''

for i in range(212):
    if (i == 0) is True:
        b = str(int(start[-2:])+1)
        next = start[0:5] + '0' + b
    else:        
        if (next[-2:] == '12') is True:
            a = str(int(next[0:4])+1)
            next = a + '-01'
        else:
            if int(next[-2:]) < 9:
                b = str(int(next[-2:])+1)
                next = next[0:5] + '0' + b
            else:
                b = str(int(next[-2:])+1)
                next = next[0:5] + b    
    dates.append(next)
```

```python
url = 'https://www.wunderground.com/history/monthly/EHAM/date/2014-02'
newurl = ''
url_lst = ['https://www.wunderground.com/history/monthly/EHAM/date/2014-02']

for i in range(len(dates)):
    newurl = url[:-7] + dates[i]
    url_lst.append(newurl)
```

```python
results_list = []
p = 0

for i in url_lst:   
    p += 1
    #Go to URL
    driver.get(i)
    # After coping the full X-path in the webpage
    sleep(5)
    my_info = driver.find_element_by_xpath("/html/body/app-root/app-history/one-column-layout/wu-header/sidenav/mat-sidenav-container/mat-sidenav-content/div/section/div[2]/div[1]/div[5]/div[1]/div/lib-city-history-observation/div/div[2]")
    print("{:.2%}".format(p/len(url_lst)), ' done')
    results_list.append(my_info.text)
```

## Data frame build up

- Web info into a list

```python
b2_lst = []

for i in results_list:
    b2 = i.split('\n')
    b2_lst.append(b2)
```

- Defining the header and where to split the list to separate the info into columns

```python
# defining the header

header = ['year_month', 'day', 'temp_F', 'dew_point_F', 'humidity', 'wind_speed_mph', 'pressure_Hg']#, 'precipitation']

# defining where to split the each list to separate the info into columns

bigcuts = []

for x in b2_lst:
    pattern = 'Max Avg Min'
    pattern2 = 'Total'
    
    cuts = [i for i, e in enumerate(x) if e == pattern or e == pattern2]
    bigcuts.append(cuts)
```

- Addressing information to each header title - creating the columns

```python
dates.insert(0, start)

y_m = []

day = []

temp_F = []

dew_point_F = []

humidity = []

wind_speed_mph = []

pressure_Hg = []

precipitation = []

for b, c, d in zip(b2_lst, bigcuts, dates):
    
    daytemp = b[2:cuts[0]]
    day.extend(daytemp)
    
    y_mtemp = [d] * len(daytemp)
    y_m.extend(y_mtemp)

    temp_Ftemp = b[cuts[0]+1:cuts[1]]
    temp_F.extend(temp_Ftemp)

    dew_point_Ftemp = b[cuts[1]+1:cuts[2]]
    dew_point_F.extend(dew_point_Ftemp)

    humiditytemp = b[cuts[2]+1:cuts[3]]
    humidity.extend(humiditytemp)

    wind_speed_mphtemp = b[cuts[3]+1:cuts[4]]
    wind_speed_mph.extend(wind_speed_mphtemp)

    pressure_Hgtemp = b[cuts[4]+1:cuts[5]]
    pressure_Hg.extend(pressure_Hgtemp)

    precipitationtemp = b[cuts[5]+1:]
    precipitation.extend(precipitationtemp)
    
```

- Creating the data frame

```python
# for the global case

bigdf = pd.DataFrame(list(zip(y_m, day, temp_F, dew_point_F, humidity, wind_speed_mph, pressure_Hg))#, precipitation))
                  , columns = header)
```

```python
bigdf.shape
```

- Cleaning info from the columns

```python
col_to_change = ['temp_F', 'dew_point_F', 'humidity', 'wind_speed_mph', 'pressure_Hg']
for i in col_to_change:
    new = bigdf[i].str.split(' ', n = 2, expand = True)
    bigdf[i] = new[1]
```
