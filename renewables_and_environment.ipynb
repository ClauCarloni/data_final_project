```python
# dutchemissions Road traffic - passenger cars (A048112)
dutchemissions_passengercars = dutchemissions[(dutchemissions.Sources == 'A048112')]
dutchemissions_passengercars
```

```python
dutchemissions_passengercars.Emissions.unique()
```

```python
dutchemissions[(dutchemissions.Emissions == 'A044108')]['EmissionsDutchTerritory_1']
```

```python
keys_list = ['A044109', 'A044108', 'A044116', 'A044117', 'A044115', 'A044107', 'A044118', 'A044112', 'A044110', 'A044114', 'A044113']

values_list = ['CO2', 'CO', 'VOC_emission_by_combustion', 'VOC_emission_by_evaporation', 'VOC', 'CH4', 'NMVOS', 'NOx', 'N2O', 'SO2', 'PM10']

zip_iterator = zip(keys_list, values_list) #Get pairs of elements

emissions_dict = dict(zip_iterator) #Convert to dictionary

print(emissions_dict)
```

```python
dutchemissions_passengercars['Emissions'].replace(emissions_dict, inplace=True)
print(dutchemissions_passengercars)
```
